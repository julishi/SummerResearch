```python
import os
from PIL import Image, ImageFilter
import glob 
import random
import shutil 
import tensorflow as tf
import json
from tensorflow.examples.tutorials.mnist import input_data

#Declare the Empty list for Classes, otherwise it will be filled with irrelevant data
img_files = []

# set input dir path
path = '/Users/julianashihadeh/Desktop/testProject/SummerResearchTutorials'

#Change current directory to Images 
os.chdir(path + '/ISIC-SubSet')

for x in glob.glob('./*'):
    img_files.append(x.split('/')[1])

for i in img_files:
    print(i)
img_files.pop(0)
for i in img_files:
    print(i)
```

```python
if not os.path.isdir(path + '/ISIC-SubSet/FilesDivided'):
    os.mkdir(path + '/ISIC-SubSet/FilesDivided')

for i in img_files: 
    print(i)
    if not os.path.isdir(path + '/ISIC-SubSet/FilesDivided/' + i):
        print("first path")
        os.mkdir(path + '/ISIC-SubSet/FilesDivided/' + i)
        if not os.path.isdir(path + '/ISIC-SubSet/FilesDivided/' + i + '/JSONFiles'):
            os.mkdir(path + '/ISIC-SubSet/FilesDivided/' + i + '/JSONFiles')
        if not os.path.isdir(path + '/ISIC-SubSet/FilesDivided/' + i + '/JPGFiles'):
            os.mkdir(path + '/ISIC-SubSet/FilesDivided/' + i + '/JPGFiles')
        if not os.path.isdir(path + '/ISIC-SubSet/FilesDivided/' + i + '/Malignant'):
            os.mkdir(path + '/ISIC-SubSet/FilesDivided/' + i + '/Malignant')
        if not os.path.isdir(path + '/ISIC-SubSet/FilesDivided/' + i + '/Benign'):
            os.mkdir(path + '/ISIC-SubSet/FilesDivided/' + i + '/Benign')
```

```python
import json
a = 0
for i in img_files:
    print(a)
    meta_data_label = []
    json_files = []
    jpg_imgs = []
    all_json_files = glob.glob(path + '/ISIC-SubSet/' + i + '/*.json')
    for x in all_json_files:
        json_files.append(x.split('/')[8])
    print("ALL JSON FILES")
    print(json_files)
    all_jpg_files = glob.glob(path + '/ISIC-SubSet/' + i + '/*.jpg')
    for x in all_jpg_files:
        jpg_imgs.append(x.split('/')[8])
    print()
    print("ALL JPG FILES")
    print(jpg_imgs) 
    #print(jpg_imgs[0])
    a = a + 1

    print('json_files')
    x = 0
    labels = 0
    for j in all_json_files:
        #print(j)
        with open(j) as f:
            print(x)
            #data = f.read()
            #output = json.load(f, dct["benign_malignant"])
            meta_data = json.load(f)
            try:
                meta_label = meta_data["meta"]["clinical"]["benign_malignant"]
                meta_data_label.append(meta_label) 
                labels = labels + 1; 
            except:
                print("skipping" + json_files[x])
            #pprint(meta_data_label)
            shutil.copy(j, path + '/ISIC-SubSet/FilesDivided/' + i + '/JSONFiles/')
            x = x + 1
    print("JSON Data extracted")
    print(meta_data_label)
    print("Total x")
    print(x)
    print("*****************")
    print("TOTAL LABELS")
    print(labels)

    print()
    print('jpg_files')
    match = 0
    nomatch = 0
    m = 0
    b = 0
    i = 0
    print(jpg_imgs)
    for j in all_jpg_files:
        
        print("The jpg_img")
        print(jpg_imgs[i].split('.')[0])
        print("The json_file")
        print(json_files[i].split('.')[0])
        #print(j)
        print("Checking match:")
        if (str(jpg_imgs[i].split('.')[0]) == str(json_files[i].split('.')[0])):
            print("MATCHING")
            match = match + 1  
            img = Image.open(j)
            new_width  = 227
            new_height = 227
            img = img.resize((new_width, new_height), Image.ANTIALIAS)
            img.save(path + '/ISIC-SubSet/FilesDivided/' + i + '/JPGFiles/' + str(jpg_imgs[x]),'jpeg')
            #img.save(path + '/ISIC-SubSet2/FilesDivided/' + i + '/JPGFiles/' + meta_data_label[x] + jpg_imgs[x] +'.jpg', 'jpeg')
            print(jpg_imgs[x])
            #print(meta_data_label[x])
            #shutil.copy(i, path + '/ISIC-SubSet/FilesDivided/UDA-2/JPGFiles/')

            if (i != labels):
                if meta_data_label[i] == 'malignant':
                    print(meta_data_label[x])
                    shutil.copy(j, path + '/ISIC-SubSet/FilesDivided/' + i + '/Malignant/')
                    print("Malignant")
                    m = m + 1

                else:
                    print(meta_data_label[i])
                    shutil.copy(j, path + '/ISIC-SubSet/FilesDivided/' + i + '/Benign/')
                    print("Benign")
                    b = b + 1
                i = i + 1
            else:
                print("Done Copying")
        else:
            print("NO MATCH")
            nomatch = nomatch + 1
       
        print("Matches")
        print(match)
        print("No matches")
        print(nomatch)
        print("Mal")
        print(m)
        print("Benign")
        print(b)
        
        print("Pre-X" + str(x))
        print(x)
        x = x + 1
        print("Post-X")
        print(x)
        print("The folder we're in right now")
        print(i)
            
```
